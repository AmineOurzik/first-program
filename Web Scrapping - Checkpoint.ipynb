```python
import requests
from bs4 import BeautifulSoup
```

```python
def analyse(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html")
    all_elements = soup.select("#mw-content-text>div>*")
    d = {}
    title = None
    for element in all_elements:
        if element.name == "h2" or element.name == "h3":
            title = element.text
            d[title] = []
        elif title == None:
            continue
        elif element.name == "p":
            d[title].append(element.text)
    print("Voici un dictionnaire de la page en texte:")
    print(d)
    l=[]
    for link in soup.find_all('a'):
        url = link.get('href', '')
        l.append(url)
    print()
    print()
    print()
    print()
    print()
    print("Voici une liste des liens de la page:")
    print(l)
```

```python
analyse("https://fr.wikipedia.org/wiki/Union_des_r%C3%A9publiques_socialistes_sovi%C3%A9tiques")
```
